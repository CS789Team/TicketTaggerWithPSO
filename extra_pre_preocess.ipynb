```python
import pandas as pd

df=pd.read_csv('clened_dataset.csv')
df.head(10)
```

```python
# stemming & remove stop words..
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer

porter=PorterStemmer()

def stemSentence(sentence):
    token_words=word_tokenize(sentence)
    token_words
    stem_sentence=[]
    for word in token_words:
        stem_sentence.append(porter.stem(word))
        stem_sentence.append(" ")
    return "".join(stem_sentence)
```

```python
df['desc']=df['desc'].apply(stemSentence)
```

```python
df
```

```python
# Import stopwords with nltk.
from nltk.corpus import stopwords
stop = stopwords.words('english')

df['desc'] = df['desc'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

```

```python
#load fasttext model to get vectors
import fasttext
import fasttext.util

ft = fasttext.load_model('fasttext/wiki.en.bin')
print(ft.get_dimension())
```

```python
from scipy.spatial.distance import cosine
import numpy as np
import pandas as pd



df_vec= pd.DataFrame(columns=['feature' + str(i) for i in range(300)])
label=[]
df_vec['label'] = label


#texVect = []
for i in range(0, len(df)):#len(my_df)
  if((i+1)%10000 == 0 ):
       print("Text %d of %d has been processed"%(i+1,len(df)))                                                                    
  embedding= ft.get_sentence_vector((df['desc'][i]))
  embedding = np.append(embedding, df['label'][i])
  newdf = pd.DataFrame(embedding.reshape(1,-1), columns=list(df_vec))
  df_vec = pd.concat([df_vec, newdf])
```

```python
from sklearn.model_selection import train_test_split

from sklearn.model_selection import cross_val_score

import numpy as np

import pandas as pd

import pyswarms as ps

import time



# use small dataset?

small = True



start = time.time()
df=df_vec

x = df.drop('label',axis =1)

y = df['label']

print("Subset loaded!")



X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20,random_state =42)

X = X_train.values

y = y_train.values




from sklearn import linear_model



# Create an instance of the classifier

# classifier = linear_model.LogisticRegression()



from sklearn.svm import SVC # for Support Vector Classification model




classifier = SVC(kernel='rbf', probability=True)

X_selected_features = X

classifier.fit(X_selected_features, y)


subset_performance_test = (classifier.predict(X_test.values) == y_test.values).mean()

print('Subset performance on test without feature selection: %.3f percent' % (100*subset_performance_test))
```
