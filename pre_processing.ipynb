```python
df['desc']=df['title'].astype(str)+' '+df['body'].astype(str)
```

```python
df['desc'] = df['desc'].str.lower() 
```

```python
import re
from bs4 import BeautifulSoup

pat1 = r'@[A-Za-z0-9_]+'#remove users
pat2 = r'https?://[^ ]+'#websites
combined_pat = r'|'.join((pat1, pat2))
www_pat = r'www.[^ ]+'#websites
eng_num_pat = '[0123456789]'#english numbers
sharta_pat = '[_]'

def remove_emoji(text):
    emoji_pattern = re.compile("["
                            u"\U0001F600-\U0001F64F"  # emoticons
                            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                            u"\U0001F680-\U0001F6FF"  # transport & map symbols
                            u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                            u"\U00002500-\U00002BEF"  # chinese char
                            u"\U00002702-\U000027B0"
                            u"\U00002702-\U000027B0"
                            u"\U000024C2-\U0001F251"
                            u"\U0001f926-\U0001f937"
                            u"\U00010000-\U0010ffff"
                            u"\u2640-\u2642" 
                            u"\u2600-\u2B55"
                            u"\u200d"
                            u"\u23cf"
                            u"\u23e9"
                            u"\u231a"
                            u"\ufe0f"  # dingbats
                            u"\u3030"
                             "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)
    return text

def clean_text(text):
  soup = BeautifulSoup(text, 'lxml')
  souped = soup.get_text()
  try:
        bom_removed = souped.decode("utf-8-sig").replace(u"\ufffd", "?")#HTML encoding didnt decode properly so we decode it again
  except:
        bom_removed = souped
  text = re.sub(combined_pat, ' ', bom_removed)
  text = re.sub(www_pat, ' ', text)
  text = re.sub(eng_num_pat, ' ', text)
  text = re.sub('[%s]' % re.escape("""!"$%&'()*+,،-–•“”./:;<=>؟?@[\]^_`{|}~»«…#‼"""), ' ', text)  
  text = re.sub('\s+', ' ', text) 
  text = remove_emoji(text) 
  text = re.sub("\d+", " ", text)
  text = re.sub('\s+', ' ', text)  
  text = re.sub(r'[^\w]', ' ', text)  
  text = re.sub(r'\b\w{1,1}\b', ' ', text)
  text = re.sub(r'\b\w{12,}\b', ' ', text)
  text = re.sub(r"http\S+", "", text)
  text = re.sub(r'[^a-zA-Z]', " ", text)  

  return text


```

```python
df['desc']=df['desc'].apply(clean_text)
```
